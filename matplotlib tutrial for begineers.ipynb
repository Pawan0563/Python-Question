Adding Legends

```python
fig=plt.figure()
```

```python
ax=fig.add_subplot(1,1,1)
```

```python
data1=np.random.randn(1000).cumsum()
data2=np.random.randn(1000).cumsum()
data3=np.random.randn(1000).cumsum()
```

```python
ax.plot(data1, 'k', label='one')
ax.plot(data2, 'b--', label='two')
ax.plot(data3, 'r.', label='three')
```

```python
ax.legend()
```

Annotations and Drawing on a Subplot

```python
import pandas as pd
```

```python
df = pd.read_csv('FB.csv')
df.head()
```

```python
df=pd.read_csv('FB.csv', 
               index_col=0, 
               parse_dates=True)
```

```python
df.head()
```

```python
c_price=df["Close"]
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
```

```python
c_price.plot(ax=ax, style='k-')
```

```python
ax.set_title(
    'Facebook stock closing prices')
```

```python
c_price.min()
```

```python
c_price["2018-12-24"]
```

```python
date="2018-12-24"
text="Minimum"
```

```python
ax.annotate(
    text, xy=(date,125), xytext=(date,150), 
    arrowprops=dict(facecolor='red'))
```
